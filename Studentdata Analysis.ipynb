```python
from pymongo import MongoClient
```

```python
client= MongoClient('mongodb://localhost:27017/')
```

```python
client.list_database_names()
```

```python
db=client ["Students"]
```

```python
db.list_collection_names()
```

```python
db.studentdata.count_documents({})
```

```python
for i in db.studentdata.find():
    print(i)
```

```python
#number of male students
db.studentdata.count_documents({'gender':'male'})
```

```python
#number of female students
db.studentdata.count_documents({'gender':'female'})
```

```python
#group performance
a=db.studentdata.aggregate([{"$group":{'_id':{'Group':"$race/ethnicity",'lunch':"$lunch"},'count':{"$count":{}}}}])
for i in a:
    print(i)
```

```python
#testwise
a=db.studentdata.aggregate([{"$group":{'_id':{'Group':"$race/ethnicity",'test':"$test preparation course"},'count':{"$count":{}}}}])
for i in a:
    print(i)
```

```python
#groupwise reading score
a=db.studentdata.aggregate([{"$group":{'_id':{'Group':"$race/ethnicity",'readingscore':"$reading score"}}}])
for i in a:
    print(i)
```

```python
a=db.studentdata.aggregate([{'$match':{'test preparation course':"completed"}}])
for i in a:
    print(i)
```

```python
for i in db.studentdata.find({},{'_id':0,'race/ethnicity':1,"lunch":1}):
    print(i)
```

```python
db.studentdata.update_many({},[{'$set':{'reading score':{'$toInt':"$reading score"}}}])
```

```python
a=db.studentdata.aggregate([{"$group":{'_id':{'Group':"$race/ethnicity",'avgreadingscore':{"$avg":"$reading score"}}}}])
for i in a:
    print(i)
```

```python
db.studentdata.update_many({},[{'$set':{'math score':{'$toInt':"$math score"}}}])
```

```python
db.studentdata.update_many({},[{'$set':{'writing score':{'$toInt':"$writing score"}}}])
```
