```python
#connecting with the mongo client
from pymongo import MongoClient
```

```python
client= MongoClient('mongodb://localhost:27017/')
```

```python
#listing data bases
client.list_database_names()
```

```python
#taking connection with one db
db=client ["foodsales"]
```

```python
#getting the name of collection inside the database
db.list_collection_names()
```

```python
db.sales.find()
```

```python
db.sales.find_one()
```

```python
#count of documents
db.sales.count_documents({})
```

```python
#total documents
for i in db.sales.find():
    print(i)
```

```python
#number of orders in east  region
db.sales.count_documents({'Region':'East'})
```

```python
##number of orders in west  region
db.sales.count_documents({'Region':'West'})
```

```python
db.sales.count_documents({'Region':'East','Product':'Carrot'})
```

```python
for i in db.sales.find({},{'_id':0}):
    print(i)
```

```python
for i in db.sales.find({},{'Product':1}):
    print(i)
```

```python
for i in db.sales.find({},{'_id':0,'Product':1}):
    print(i)
```

```python
#Productwise and category wise count
a=db.sales.aggregate([{"$group":{'_id':{'p':"$Product",'c':"$Category"},'count':{"$count":{}}}}])
for i in a:
    print(i)
```

```python
a=db.sales.aggregate([{"$group":{'_id':{'region':"$Region",'product':"$Product"},'count':{"$count":{}}}}])
for i in a:
    print(i)
```

```python
a=db.sales.aggregate([{'$match':{'OrderDate':"06-02-2020"}}])
for i in a:
    print(i)
```

```python
db.sales.update_many({},[{'$set':{'TotalPrice':{'$toDouble':"$TotalPrice"}}}])
```

```python
db.sales.update_many({},[{'$set':{'UnitPrice':{'$toDouble':"$UnitPrice"}}}])
```

```python
for i in db.sales.find():
    print(i)
```

```python
db.sales.update_many({},[{'$set':{'Quantity':{'$toInt':"$Quantity"}}}])
```

```python
for i in db.sales.find():
    print(i)
```

```python
db.sales.aggregate([{'$group':{'_id':'$Category','sum':{'$sum':'$Quantity'}}}])
for i in db.sales.aggregate([{'$group':{'_id':'$Category','sum':{'$sum':'$Quantity'}}}]):
    print(i)
```
